## Problem 1
Convert the following data into tidy format with the headers `country`, `drink` and `servings`. Drop the column `total_litres_of_pure_alcohol`

```python
import pandas as pd

url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/alcohol-consumption/drinks.csv"
df = pd.read_csv(url)

# Drop total alcohol column
df = df.drop(columns=['total_litres_of_pure_alcohol'])

# Convert to tidy format
tidy = df.melt(id_vars='country',
               var_name='drink',
               value_name='servings')

tidy.head()
```

## Problem 2
Remove `"_servings"` from the data in the tidy dataframe in Problem 1. Who drinks the most beer, wine and spirits? Create three categorical plots of your choice to illustrate the top 10 countries that drink the most wine, beer or spirits.

```python
# Remove '_servings'
tidy['drink'] = tidy['drink'].str.replace('_servings', '')

# Top 10 for each drink
import seaborn as sns
import matplotlib.pyplot as plt

top_beer = tidy[tidy['drink']=='beer'].sort_values('servings', ascending=False).head(10)
top_wine = tidy[tidy['drink']=='wine'].sort_values('servings', ascending=False).head(10)
top_spirits = tidy[tidy['drink']=='spirit'].sort_values('servings', ascending=False).head(10)

# Plot
sns.catplot(data=top_beer, x='servings', y='country', kind='bar')
plt.title("Top 10 Beer Consumption")

sns.catplot(data=top_wine, x='servings', y='country', kind='bar')
plt.title("Top 10 Wine Consumption")

sns.catplot(data=top_spirits, x='servings', y='country', kind='bar')
plt.title("Top 10 Spirits Consumption")
```

## Problem 3
Create a wide dataframe using `pivot_table` with the tidy dataframe from Problem 2

```python
# Pivot to wide format
wide = tidy.pivot_table(index='country',
                        columns='drink',
                        values='servings')

wide.head()
```
